Checking which class has the highest survival rate :-

```python
df['Pclass'].sum()
```

```python
num_df = df[['Pclass', 'Survived']].groupby('Pclass').mean().reset_index()
print(num_df)
```

```python
num_df[['Survived']] = num_df[['Survived']] * 100
```

```python
num_df.rename(columns = {"Survived": "Survival in %"}, inplace=True)
```

```python
sns.barplot(data = num_df, x = 'Pclass', y = 'Survival in %')
```
